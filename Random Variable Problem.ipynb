```python
import random 
import matplotlib.pyplot as plt
import numpy as np 
import matplotlib.mlab as mlab 
import math as math

#Creating a data set that sums a number of random variables between 0 and 1 
nbRandVar = 200
nbRounds = 2000
sumArray=np.zeros(nbRounds)

#Sums specific number of random variables between 0 and 1
for j in range(nbRounds):
    val = 0 
    for i in range(nbRandVar):
        val = sumArray[j] 
        sumArray[j] = val + np.random.random_sample(1)

#Plotting the Histogram
plt.hist(sumArray, bins=50, color='red')
plt.xlabel('Sum of Random Variables', size = 15)
plt.ylabel('Quantity of Sums at X Value', size = 15)
plt.show()
```

```python
#Finding the mean of the distribution
testVal=0
for i in range(nbRounds): 
    testVal= testVal + sumArray[i] 

distribMean = testVal/nbRounds 
```

```python
#Comparing the data set with a normal distribution curve
mu = distribMean
variance = np.var(sumArray)
sigma = math.sqrt(variance)
x = np.linspace(mu-1.25*variance,mu+1.25*variance, 100) 
plt.plot(x,mlab.normpdf(x, mu, sigma), color = 'blue') 
plt.hist(sumArray, bins=50, color='red', normed = True)
plt.xlabel('Sum of Random Variables', size = 15)
plt.ylabel('Fraction of Sums at X Value', size = 15)
plt.show() 
plt.show()
```
