```python
import random 
import matplotlib.pyplot as plt
import numpy as np 
import matplotlib.mlab as mlab

#seed=2600
#np.random.seed(seed)

#Number of Coins per set of tosses and number of tosses the coin(s) are tossed
nbCoins = 1000
nbDraws = 10000
#Probability for heads is 50%
probaHeads = .5
resultsArray = np.empty(nbDraws)

#Creates coin toss data and puts it in array
for j in range(nbDraws):
    nbHeads=0
    for i in range(nbCoins): 
        val = np.random.random_sample()
        if val<probaHeads: 
            nbHeads = nbHeads + 1 
    resultsArray[j] = nbHeads
    
#plots histogram        
plt.hist(resultsArray, bins=100, color='orange') 
plt.xlabel('Sum of Heads', size=20)
plt.ylabel('Number of Draws', size=20)
plt.show()
```

```python
#Zoomed in plot of the previous histogram
ax1 = plt.subplot()         
plt.hist(resultsArray, bins=100, color='orange') 
plt.xlabel('Sum of Heads', size=20)
plt.ylabel('Number of Draws', size=20)
ax1.set_xlim([480, 520])
plt.show()
```

```python
#comparing the coin toss results with a binomial distribution
s = np.random.binomial(nbCoins, probaHeads, nbDraws)
plt.hist(s, bins=100) 
plt.hist(resultsArray, bins=100, color='orange') 
plt.xlabel('Sum of Heads', size=20)
plt.ylabel('Number of Draws', size=20)
plt.show()
plt.show()
```
